```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import plotly.express as px
import plotly.io as pio
import matplotlib.image as mpimg
import plotly.graph_objects as go
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('/content/crimes_against_women_2001-2014.csv')
data.head()
```

```python
data=data.drop(['Unnamed: 0'],axis=1)
data.info()
```

```python
data.isna().sum()/len(data)*100
```

```python
data['STATE/UT'].unique()
```

```python
data.columns

```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(),annot=True,cmap='cubehelix',fmt='.2f',linewidths=1)
plt.show()
```

```python
plt.figure(figsize=(10,7))
data['Year'].value_counts().plot(kind='line',color='red')
plt.title("Visualize the Years in the data")
plt.show()
```

```python
data.groupby('Year')['Rape'].sum()\
.plot(kind='line',color='black',figsize=(10,8))
plt.title("The Sum of the Rape cases with year wise")
plt.xlabel("Year")
plt.ylabel("Sum of the Rape cases")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=data,x=data['STATE/UT'])
plt.title("Visualize the States in the data")
plt.xticks(rotation=90)
plt.show()
```

```python
data.groupby('STATE/UT')['Rape'].sum().sort_values(ascending=False)\
.plot(kind='bar',color=['#2ECCFA','#9AFE2E','#FA58D0','#2EFEC8','#F4FA58','#170B3B'],figsize=(12,8))
plt.title("How Many Rape cases recorded in Statewise")
plt.xlabel("Sates")
plt.ylabel("The Sum of the rape cases")
plt.show()
```

Observations:
From the data covering the years 2001 to 2014, we have discovered several interesting facts and trends.
1. The countplot chart reveals that Uttar Pradesh and Madhya Pradesh have the highest number of recorded values compared to other states.
2. The line chart depicting the crime rate over the years shows a consistent increase in crime rates year after year.
3. The line chart representing the recorded rape cases clearly indicates a significant rise in cases each year, suggesting a concerning trend.
4. According to the bar chart, Madhya Pradesh tops the list in terms of the highest number of recorded rape cases, followed by Bihar in the second position, indicating a higher incidence of these cases in these states.
5. Analyzing the data on dowry cases, Uttar Pradesh emerges as the state with the highest number of recorded cases from 2001 to 2014, emphasizing the prevalence of this issue in the region.

```python
plt.figure(figsize=(10,8))
sns.histplot(data,x='Assault on women with intent to outrage her modesty',bins=50,kde=True,color='red')
plt.title("Visualize the Assault on women with intent to outrage her modesty in the data")
plt.show()
```

```python
kidnapp_cases=data[['Year','Kidnapping and Abduction']]
kidnapp_cases.groupby('Year')['Kidnapping and Abduction'].sum()\
.plot(kind='line',color='green',figsize=(10,8))
plt.title("Which Year most Kindnapping and abductions cases recorded in 2001 to 2014")
plt.xlabel("Year")
plt.ylabel("Sum of the Kidnapping and abduction cases")
plt.show()
```

```python
Andhra_pradesh=data[data['STATE/UT']=='ANDHRA PRADESH']
rape_cases_in_andhra=Andhra_pradesh.groupby('Year')['Rape'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,8))
sns.scatterplot(data=rape_cases_in_andhra,x=rape_cases_in_andhra.index,y=rape_cases_in_andhra.values,color='red')
plt.xlabel("Year")
plt.ylabel("Sum of the Rape Cases")
plt.title("Find the Rape cases recorded in Andhra Pradesh Sate")
plt.show()
```

```python
Dowry_death=Andhra_pradesh.groupby('Year')['Dowry Deaths'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,6))
sns.scatterplot(data=Dowry_death, x=Dowry_death.index, y=Dowry_death.values)
plt.show()
```

```python
Utterpradesh=data[data['STATE/UT']=='UTTAR PRADESH']
Utterpradesh.groupby('Year')['Cruelty by Husband or his Relatives'].sum().sort_values(ascending=False)\
.plot(kind='pie',title="Total Cruelty by Husband or his Relatives Cases Recorded in Uttar Pradesh",autopct='%1.2f%%',
    pctdistance=0.8,
    shadow=True,
    labeldistance=1.1)
```
